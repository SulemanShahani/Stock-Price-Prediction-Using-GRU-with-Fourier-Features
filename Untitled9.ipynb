```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, GRU, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Function to calculate additional evaluation metrics
def calculate_metrics(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    mfe = np.mean(y_true - y_pred)
    mad = np.mean(np.abs(y_true - y_pred))
    r2 = r2_score(y_true, y_pred)
    return mae, mape, mfe, mad, r2

# Ticker symbol for the stock
ticker_symbol = 'AAPL'

# Fetch data from Yahoo Finance
stock_data = yf.download(ticker_symbol, start='2020-01-01', end='2024-01-01', progress=False)

# Function to create Fourier features
def create_fourier_features(df, fourier_terms=4):
    time = np.arange(len(df))
    T = len(df)

    for i in range(1, fourier_terms + 1):
        df[f'fourier_sin_{i}'] = np.sin(2 * np.pi * i * time / T)
        df[f'fourier_cos_{i}'] = np.cos(2 * np.pi * i * time / T)

    return df

# Create Fourier features
stock_data = create_fourier_features(stock_data, fourier_terms=4)

# Drop rows with NaN values due to the creation of Fourier features
stock_data.dropna(inplace=True)

# Define features
FEATURES = ['fourier_sin_1', 'fourier_cos_1', 'fourier_sin_2', 'fourier_cos_2',
            'fourier_sin_3', 'fourier_cos_3', 'fourier_sin_4', 'fourier_cos_4']

# Scale the 'Close' price
scaler_close = MinMaxScaler(feature_range=(0, 1))
scaled_close = scaler_close.fit_transform(stock_data[['Close']])

# Scale the features
scaler_features = MinMaxScaler(feature_range=(0, 1))
scaled_features = scaler_features.fit_transform(stock_data[FEATURES])

# Combine scaled 'Close' price and scaled features
scaled_data = np.concatenate([scaled_close, scaled_features], axis=1)

# Prepare the data for LSTM/GRU
def create_sequences(data, sequence_length):
    sequences = []
    labels = []
    for i in range(len(data) - sequence_length):
        seq = data[i:i + sequence_length]
        label = data[i + sequence_length, 0]  # 'Close' price
        sequences.append(seq)
        labels.append(label)
    return np.array(sequences), np.array(labels)

sequence_length = 100  # Increased sequence length
X, y = create_sequences(scaled_data, sequence_length)

# Split the data into training and testing sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Random search for hyperparameter tuning
best_rmse = float('inf')
best_params = None
best_model = None

# Define ranges for random search
units_options = [50, 100, 150]
batch_size_options = [16, 32, 64]
learning_rate_options = [0.001, 0.01, 0.05]
dropout_rate_options = [0.2, 0.3, 0.4]

for units in units_options:
    for batch_size in batch_size_options:
        for lr in learning_rate_options:
            for dropout_rate in dropout_rate_options:
                # Build GRU model
                # Build GRU model with dropout regularization
                model = Sequential([
                    GRU(units, return_sequences=True, input_shape=(sequence_length, X.shape[2])),
                    Dropout(dropout_rate),  # Add dropout regularization
                    GRU(units),
                    Dropout(dropout_rate),  # Add dropout regularization
                    Dense(1)
                ])


                model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=lr), loss='mean_squared_error')

                # Early stopping
                early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

                # Train GRU model
                history = model.fit(X_train, y_train, epochs=100, batch_size=batch_size, validation_data=(X_test, y_test), verbose=0, callbacks=[early_stopping])

                # Predict with GRU model
                pred = model.predict(X_test)
                pred = scaler_close.inverse_transform(pred)  # Reverse scaling

                # Remove NaNs for RMSE calculation
                y_test_nonan = y_test[~np.isnan(y_test)]
                pred_nonan = pred[~np.isnan(y_test)]

                if len(y_test_nonan) == 0 or len(pred_nonan) == 0:
                    continue




                # Evaluate GRU model
                rmse = np.sqrt(mean_squared_error(scaler_close.inverse_transform(y_test.reshape(-1, 1)), pred))
                if rmse < best_rmse:
                    best_rmse = rmse
                    best_params = (units, batch_size, lr, dropout_rate)
                    best_model = model

print(f'Best RMSE: {best_rmse:.2f} with units: {best_params[0]}, batch size: {best_params[1]}, learning rate: {best_params[2]}, dropout rate: {best_params[3]}')

# Save the best model
best_model.save('best_gru_model.h5')

# Predict and evaluate with the best model
best_pred = best_model.predict(X_test)
best_pred = scaler_close.inverse_transform(best_pred)  # Reverse scaling
best_mae, best_mape, best_mfe, best_mad, best_r2 = calculate_metrics(scaler_close.inverse_transform(y_test.reshape(-1, 1)), best_pred)

print(f'GRU - RMSE: {best_rmse:.2f}, MAE: {best_mae:.2f}, MAPE: {best_mape:.2f}%, MFE: {best_mfe:.2f}, MAD: {best_mad:.2f}, R2: {best_r2:.2f}')

# Save the best model
best_model.save('best_gru_model.h5')

# Plot actual vs predicted prices for the best model
plt.figure(figsize=(14, 7))
plt.plot(stock_data.index[-len(y_test):], scaler_close.inverse_transform(y_test.reshape(-1, 1)), label='Actual', color='blue')
plt.plot(stock_data.index[-len(best_pred):], best_pred, label='Predicted', color='red')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title(f'Actual vs Predicted Stock Prices for {ticker_symbol} - GRU (Best Model)')
plt.legend()
plt.grid(True)
plt.show()
```
